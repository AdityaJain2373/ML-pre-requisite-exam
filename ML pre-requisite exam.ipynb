Question 2

```python
import numpy as np
from time import time

M1 = np.random.rand(2,3)*4
M2 = np.random.rand(3,4)*4

n = len(M1)
k = len(M1[0])
m = len(M2[0])

res_M = [[0 for _ in range(m)] for _ in range(n)]

for row in range(n):
    for col in range(m):
        element = 0
        for i in range(k):
            element += M1[row][i] * M2[i][col]
        res_M[row][col] = element

s = time()
print(f'Manual multiply took {round((time()-s)*1000,5)} ms')
print(np.array(res_M))

print()
s = time()
numpy_answer = np.matmul(M1,M2)
print(f'Numpy multiply took {round((time()-s)*1000,5)} ms')
print(numpy_answer)
```

Question 4 - Gradient of the function f(x) = (x^2)*y + (y^3)*sin(x)

Answer - Gradient with respect to x =  2*x*y + (y^3)*cos(x)
         Gradient with respect to y =  x^2 + 3*(y^2)sin(x)

Question 5 - Use JAX to confirm the gradient

```python
import jax
import jax.numpy as jnp

def f(x,y):
    return (x**2) * y + (y**3) * jnp.sin(x)

gradient = jax.grad(f, argnums = (0,1))

x_val = 1.0
y_val = 2.0

grad_x ,grad_y = gradient(x_val, y_val)

print(grad_x)
print(grad_y)
```

Question 6 - Use sympy to confirm the gradient

```python
import sympy as sp

x,y = sp.symbols('x y')

f = (x**2) * y + (y**3) * sp.sin(x)

print(sp.diff(f,x))
print(sp.diff(f,y))
```

Question 8 - Do the previous question using classes and inheritence

```python
class Student():
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def __repr__(self):
        return f"Student(Name: {self.name}, Roll Number: {self.roll_number}, Marks: {self.marks})"

class Branch():
    def __init__(self,branch):
        self.branch = branch
        self.students = []

    def add_student(self,student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch(Name: {self.branch}, Students: {self.students})"

class Year():
    def __init__(self,year):
        self.year = year
        self.branches = []

    def add_branch(self,branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"Year(Year: {self.year}, Branches: {self.branches})"

class DataBase:
    def __init__(self):
        self.years = []

    def add_year(self,year):
        self.years.append(year)

    def __repr__(self):
    
        return f"DataBase(Years: {self.years})"


student1 = Student('N',1, {'Maths' : 100, 'English' : 70})

branch1 = Branch('Branch 1')
branch1.add_student(student1)
branch2 = Branch('Branch 2')

year2022 = Year(2022)
year2022.add_branch(branch1)
year2022.add_branch(branch2)

database = DataBase()
database.add_year(year2022)

print(database)
```

Question 12

Broadcasting is the ability of python (numpy) of using arrays of different dimensions during arithmetic operations.

```python
import numpy as np

N1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
N2 = np.array([1,2,3])

N = N1 + N2
print(N)                                               # Here both arrays had different dimensions but numpy treated N2 as 3 cross 3 array.
print()

N_ = N1 * N2
print(N_)                                              # Here, numpy did multiplication of a (3,3) array with a (1,3) array.
```
